```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def datascraping(url):
    url='https://meesho.com/bags-ladies/pl/p7vbp'
    page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
    soup=BeautifulSoup(page.content)
    names= []
    for i in soup .find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
        names.append(i.text)
    price=[]
    for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
        price.append(i.text)
    discount=[]
    for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
        discount.append(i.text)
    df=pd.DataFrame({'ProductNames':names,'Price':price,'Discounts':discount})
    return df
```

```python
def datascraping(url):
    url='https://www.dineout.co.in/delhi-restaurants/buffet-special'
    page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
    soup=BeautifulSoup(page.content)
    name= []
    for i in soup .find_all('a',class_="restnt-name ellipsis"):
        name.append(i.text)
    cuisine= []
    for i in soup .find_all('span',class_="double-line-ellipsis"):
        cuisine.append(i.text)
    location= []
    for i in soup .find_all('div',class_="restnt-loc ellipsis"):
        location.append(i.text)
    rating= []
    for i in soup .find_all('div',class_="restnt-rating rating-4"):
        rating.append(i.text)
    images= []
    for i in soup .find_all('img',class_="no-img"):
        images.append(i['data-src'])
    df=pd.DataFrame({'Name':name,'Cuisine':cuisine,'Location':location,'Rating':rating,'Images':images})
    return df
```

```python
def datascraping(url):
    url='https://coreyms.com/'
    page=requests.get('https://coreyms.com/')
    soup=BeautifulSoup(page.content)
    heading=[]
    for i in soup .find_all('a',class_="entry-title-link"):
        heading.append(i.text)
    date=[]
    for i in soup .find_all('time',class_="entry-time"):
        date.append(i.text)
    content=[]
    for i in soup .find_all('div',class_="entry-content"):
        content.append(i.text)
    videos=[]
    for i in soup .find_all('iframe',class_="youtube-player"):
        videos.append(i['src'])
    videos.insert(4,0)
    df=pd.DataFrame({'Heading':heading,'Date':date,'Content':content,'VideoURLS':videos})
    return df
```

```python
def datascraping(url):
    url='https://www.imdb.com/india/top-rated-indian-movies/'
    page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
    soup=BeautifulSoup(page.content)
    name_year_of_release=[]
    for i in soup .find_all('td',class_="titleColumn"):
        name_year_of_release.append(i.text)
    names_year=name_year_of_release[0:100]
    rating=[]
    for i in soup .find_all('td',class_="ratingColumn imdbRating"):
        rating.append(i.text)
    ratings=rating[0:100]
    df=pd.DataFrame({'Names & Year of Yelease':names_year,'Ratings':ratings})
    return df
```

```python
def datascraping(url):
    url='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
    page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
    soup=BeautifulSoup(page.content)
    team=[]
    for i in soup .find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    teams=team[0:10]
    match=[]
    for i in soup .find_all('td',class_="rankings-block__banner--matches"):
        match.append(i.text)
    point=[]
    for i in soup .find_all('td',class_="rankings-block__banner--points"):
        point.append(i.text)
    rating=[]
    for i in soup .find_all('td',class_="rankings-block__banner--rating u-text-right"):
        rating.append(i.text.replace('\n',''))
    matches=[]
    for i in soup .find_all('td',class_="table-body__cell u-center-text"):
        matches.append(i.text)
    mat=matches[0::2]
    mat.insert(0,match)
    mat=mat[0:10]
    pt=matches[1::2]
    pt.insert(0,point)
    pt=pt[0:10]
    ratings=[]
    for i in soup .find_all('td',class_="table-body__cell u-text-right rating"):
        ratings.append(i.text)
    ratings.insert(0,rating)
    ratings=ratings[0:10]
    df=pd.DataFrame({'Teams':teams,'Matches':mat,'Points':pt,'Ratings':ratings})
    return df
```
